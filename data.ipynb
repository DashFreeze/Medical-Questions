```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Diseases & treatments questions.csv")

df.head()
```

```python
df = df.iloc[:, 1:]    
```

```python
relevant_values = {"Helpful", "Unhelpful"}

for column in df.columns:
    if df[column].isin(relevant_values).any():
        counts = df[column].value_counts()
        explode = [0.1 if i == 0 else 0 for i in range(len(counts))]

        plt.figure(figsize=(6, 6))
        counts.plot(kind="pie", 
                    autopct='%1.1f%%',
                    colors=["skyblue", "orange"],
                    startangle=90,
                    explode=explode)
        plt.title(column)
        plt.ylabel("")
        plt.show()
    else:
        counts = df[column].value_counts(normalize=True) * 100
        counts = counts.reindex(range(1, 6), fill_value=0)

        plt.figure(figsize=(8, 5))
        counts.plot(kind="bar", color="skyblue", width=0.8)

        plt.title(column)
        plt.xlabel("")
        plt.ylabel("Häufigkeit (%)")
        plt.ylim(0, max(counts.max() + 5, 100))

        plt.xticks(range(5), labels=range(1, 6))

        for i, value in enumerate(counts):
            plt.text(i, value + 1, f"{value:.1f}%", ha="center")

        plt.show()
```

```python
columns_to_drop = df.columns[::3]
df = df.drop(columns=columns_to_drop)

average_values = df.mean()

sorted_columns = average_values.sort_values(ascending=True).index
sorted_df = df[sorted_columns]

sorted_df.head()
```
